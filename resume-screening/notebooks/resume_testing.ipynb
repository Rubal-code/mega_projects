```python
import joblib

model = joblib.load("../model/resume_model.pkl")   
vectorizer = joblib.load("../model/tfidf.pkl")
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', ' ', text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = text.split()
    words = [
        lemmatizer.lemmatize(w)
        for w in words
        if w not in stop_words and len(w) > 2
    ]
    return ' '.join(words)
```

```python
test_resume = "Python developer with experience in machine learning and data analysis"

cleaned = clean_text(test_resume)
cleaned
```

```python
resume_text = """
Experienced Java developer with strong knowledge of Spring Boot,
REST APIs, MySQL, Hibernate, and backend development.
"""

resume_clean = clean_text(resume_text)
resume_vec = vectorizer.transform([resume_clean])

predicted_category = model.predict(resume_vec)
predicted_category[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

job_description = """
Looking for a backend Java developer skilled in Spring Boot,
REST APIs, MySQL, and microservices.
"""

jd_clean = clean_text(job_description)
jd_vec = vectorizer.transform([jd_clean])

fit_score = cosine_similarity(resume_vec, jd_vec)[0][0] * 100
round(fit_score, 2)
```
